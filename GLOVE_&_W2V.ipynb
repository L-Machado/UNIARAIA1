```python

import gensim.downloader as api

# Carregar embeddings GloVe pré-treinados
glove_model = api.load("glove-wiki-gigaword-100")

# Obter vetor de palavra para uma palavra específica
word_vector = glove_model['dog']
print("Vetor GloVe para a palavra 'dog':", word_vector)

# Encontrar palavras mais similares
similar_words = glove_model.most_similar('dog', topn=5)
print("Palavras mais similares a 'dog':", similar_words)
```

```python
import gensim.downloader as api

# Carregar embeddings Word2Vec pré-treinados
w2v_model = api.load("word2vec-google-news-300")

# Obter vetor de palavra para uma palavra específica
word_vector = w2v_model['dog']
print("Vetor Word2Vec para a palavra 'cachorro':", word_vector)

# Encontrar palavras mais similares
similar_words = w2v_model.most_similar('dog', topn=3)
print("Palavras mais similares a 'dog':", similar_words)
```
